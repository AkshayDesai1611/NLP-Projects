```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
classifier("He is a skilled employee")
```

```python
results = classifier(["We are really excited to build chatbots using langchain.",
            "However there might be restriction on the number of tokens to be used"])
```

```python
for result in results:
  print(f"label: {result['label']},with score: {round(result['score'],4)}")
```

```python
from transformers import AutoTokenizer,TFAutoModelForSequenceClassification
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = TFAutoModelForSequenceClassification.from_pretrained(model_name,from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline("sentiment-analysis",model=model,tokenizer=tokenizer)
```

```python
classifier("I am great but not a wealthy guy")
```
